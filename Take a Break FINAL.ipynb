```python

# Define global variables
global ply_list, history, breaks, store_list, shuf_stat, count

# Importing Modules
from tkinter import *
import tkinter.font as font
import webbrowser
from tkinter import messagebox
import time as tm
from time import strftime
import random
import datetime

#Define the window and its properties
root = Tk()
root.title("Python GUI")
root.resizable(False, False)
root.geometry('700x600')
root.configure(bg='lightblue')


# All Functions

# 1. Start the break function


def start():
    global ply_list, history, breaks, shuf_stat, store_list, count, snz_time
    
    if breaks == 0:
        
        cr_tm = datetime.datetime.now()
        cr_tm10 = cr_tm + datetime.timedelta(seconds= 30 + snz_time) 
        cr_tm10 = cr_tm10.strftime('%I:%M:%S %p')
        
        lbl_nxtbrktm.configure(text= 'Next Video plays at: \n{}'.format(cr_tm10))
        
        messagebox.showinfo('Info','Take a Break program started!')
        

    while shuf_stat == 1:        
                
        ply_list = store_list
        
        shuf_stat = 0

    while breaks < len(ply_list):
        
        tm.sleep(20)
        
        msg_stat = messagebox.askyesno('Do you want to continue?','If Yes...\n Break {} starts in {} secs!'.format(breaks+1,10+snz_time))
        msg_stat = str(msg_stat)
        
        
        if msg_stat == 'False':
            
            break
        
        else:
                                             
            tm.sleep(10+snz_time)

            webbrowser.open(str(ply_list[count]['site']))

            history[count] = ply_list[count]

            lbox_hist.insert(count,'Video {}: {}'.format(count+1,history[count]['name']))

            count +=1

            tm.sleep(10)
            
            cr_tm = datetime.datetime.now()
            cr_tm10 = cr_tm + datetime.timedelta(seconds = 30 + snz_time)
            cr_tm10 = cr_tm10.strftime('%I:%M:%S %p')

            lbl_nxtbrktm.configure(text= 'Next Video plays at: \n{}'.format(cr_tm10))

            messagebox.showinfo('Info','Break time over, Get back to work')
            
            breaks +=1
            snz_time = 0

# 2. End the program

def exit():
    root.destroy()

# 3. Displaying system time
    
def clock(): 
    string = strftime('%I:%M:%S %p') 
    lbl_clock.config(text = 'Current time: '+ string) 
    lbl_clock.after(1000, clock) 

# 4. Shuffle

def shuffle():
    global shuf_stat
    shuf_stat = 1
    
    lbox_play.delete(0,END)
    
    order = list(ply_list.keys())
    random.shuffle(order)
     
    list_items = len(ply_list)
    
    for i in range(0,list_items):
        
        lbox_play.insert(i+1,'Video {}: {}'.format(i+1,ply_list[order[i]]['name']))
        store_list[i] = ply_list[order[i]]
      
        

# 5. Skip to Next Video:
def skip():        
    global count
    count +=1
    messagebox.showinfo('Info','Next Video skipped.')
    
# 6. Skip to Next Video:
def snooze():
    global snz_time
    snz_time = 5
    messagebox.showinfo('Info','Video snoozed for {} seconds'.format(snz_time))

# Defining global variables


history = {}
store_list = {}
breaks = 0
shuf_stat = 0
snz_time = 0
count = 0

ply_list = {
            0:{'name':'From Engineering Room To F1 Cockpit','site':'https://www.youtube.com/watch?v=Nkea6ltQyY8'}, 
            1:{'name':'The Insane Logistics Of Formula 1','site':'https://www.youtube.com/watch?v=6OLVFa8YRfM'}, 
            2:{'name':'Top 10 Cheeky F1 Innovations','site':'https://www.youtube.com/watch?v=rnIjQC08qKk'}, 
            3:{'name':"Sangakkara Gets First Lord's Hundred",'site':'https://www.youtube.com/watch?v=GwN0zLotgqc'}, 
            4:{'name':'TED - Lessons Learned From India','site':'https://www.youtube.com/watch?v=HRBzRphZ3rQ'},
            5:{'name':'DC Shoes: Ultimate Urban Playground','site':'https://www.youtube.com/watch?v=LuDN2bCIyus'}, 
            6:{'name':"Malaysia's $100bn Smart Island City",'site':'https://www.youtube.com/watch?v=QFG-PIJ0GUE'}, 
            7:{'name':'Top 10 Best Races Of The Decade','site':'https://www.youtube.com/watch?v=1TOVYKbjwOk'}, 
            8:{'name':'How American C.E.Os. Got So Rich','site':'https://www.youtube.com/watch?v=ylLTMYt24lA'}, 
            9:{'name':'Charity Match betweeen Cricket Legends','site':'https://www.youtube.com/watch?v=Vk3DwEcy_70'}
            }


lblfont = font.Font(family='Verdana', size= 20, weight='bold')
lblfont2 = font.Font(family='Verdana', size= 18)
lblfont3 = font.Font(family='Verdana', size= 14)
txtfont = font.Font(family='Verdana', size= 10)
btnfont = font.Font(family='Verdana', size= 10)



# Placing Widgets on GUI:

lbl_clock = Label(root, font = lblfont2 , background = 'lightblue', foreground = 'black') 
lbl_clock.grid(row=0,column=10, columnspan=20)

lbl_ply = Label(root, text="Playlist", font=lblfont, bg='lightblue')
lbl_ply.grid(row=2, column=5)

lbox_play = Listbox(root, selectmode = SINGLE, font = txtfont, bg='lightblue', height=10, width=45)
lbox_play.grid(row = 3, column = 0, columnspan = 15, rowspan=2, padx=10, pady=15)


photo = Canvas(width=300, height=250, bg='lightblue', highlightthickness=0)
file = PhotoImage(file = "coffee-break2.png")
photo.create_image(150,125, image=file)
photo.grid(row =1, column = 15, rowspan=5, columnspan=10,padx=25, pady=15 )

lbl_head = Label(root, text="Take a Break", font=lblfont, bg='lightblue')
lbl_head.grid(row=5, column=12, columnspan=12,padx=25, pady=15)

btn_strt = Button(root, text="START",font = btnfont, command= start)
btn_strt.grid(row=5, column=0)

btn_strt = Button(root, text="SKIP NEXT",font = btnfont, command= skip)                    
btn_strt.grid(row=5, column=3)  

btn_strt = Button(root, text="SNOOZE",font = btnfont, command= snooze)                    
btn_strt.grid(row=5, column=5) 

btn_ext = Button(root, text="EXIT",font = btnfont, command= exit)
btn_ext.grid(row=5, column=6)

btn_shfl = Button(root, text="SHUFFLE",font = btnfont, command=shuffle)
btn_shfl.grid(row=5, column=10)


lbl_hist = Label(root, text="History", font=lblfont, bg='lightblue')
lbl_hist.grid(row=8, column=5,rowspan=2)

lbox_hist = Listbox(root, selectmode = SINGLE, font = txtfont, bg='lightblue',  width=45)
lbox_hist.grid(row = 10, column = 0, columnspan = 15, rowspan=2, padx=10, pady=15)

lbl_nxtbrktm = Label(root, font=lblfont2, bg='lightblue')
lbl_nxtbrktm.grid(row=10, column=12, columnspan=12,padx=25, pady=15)

# Inserting video links in Playlist

list_items = len(ply_list)

for i in range(0,list_items):
    lbox_play.insert(i+1,'Video {}: {}'.format(i+1,ply_list[i]['name']))
    
clock()
mainloop()
```
